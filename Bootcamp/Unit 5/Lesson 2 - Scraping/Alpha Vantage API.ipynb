# Plotting Data

```python
from alpha_vantage.timeseries import TimeSeries

# Plots time series data for Microsoft Stock (MSFT)
ts = TimeSeries(key, output_format='pandas')
data, meta_data = ts.get_intraday(symbol='MSFT',interval='15min', outputsize='full')
data['4. close'].plot()
plt.title('Intraday Times Series for the MSFT stock (15 min)')
plt.show()
```

```python
from alpha_vantage.techindicators import TechIndicators

# Views technical indicator - Bolliger Bands
ti = TechIndicators(key, output_format='pandas')
data, meta_data = ti.get_bbands(symbol='MSFT', interval='60min', time_period=60)
data.plot()
plt.title('BBbands indicator for  MSFT stock (60 min)')
plt.show()
```

```python
from alpha_vantage.sectorperformance import SectorPerformances

# View sector performance
sp = SectorPerformances(key, output_format='pandas')
data, meta_data = sp.get_sector()
data['Rank A: Real-Time Performance'].plot(kind='bar')
plt.title('Real Time Performance (%) per Sector')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
from alpha_vantage.cryptocurrencies import CryptoCurrencies

# View cryptocurrency data, because why not?
cc = CryptoCurrencies(key, output_format='pandas')
data, meta_data = cc.get_digital_currency_intraday(symbol='BTC', market='CNY')
data['1b. price (USD)'].plot()
plt.tight_layout()
plt.title('Intraday value for bitcoin (BTC)')
plt.grid()
plt.show()
```
