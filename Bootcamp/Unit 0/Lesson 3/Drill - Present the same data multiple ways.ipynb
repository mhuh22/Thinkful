# Plotting market value

```python
df2 = pd.read_csv('thads2013n.txt')
df3 = df2[['BUILT',
                               'FMR',
                               'FMTBEDRMS',
                               'FMTOWNRENT',
                               'FMTREGION',
                               'PER',
                               'TOTSAL', 
                               'VALUE',
                               'ZINC2',
                               'ZSMHC']]
df3.head()
```

```python
sns.set_style("whitegrid")
ax = sns.boxplot(y=df3['FMR'], data=df3)
```

Pros

    Easy to find median, quartiles, and outliers
    Shows where most values lie between
    Can show distribution skew with the number of outliers
    
Cons

    Hard to visualize outliers if there are a lot of them
    Doesn't show population size

```python
sns.violinplot(df3['FMR'], vert=False)
```

Pros

    Shows where most common values lie
    Shows median and quartiles
Cons

    Doesn't include population size or distribution
    Can't see minimum and maximum values

```python
sns.distplot(df3['FMR'], kde=False)
```

Pros 

    Best visualization for showing distribution 
    Shows most common values, and outliers
    Shows quantity of units for each price
    
Cons

    Hard to visualize the minimum and maximum values

```python
sns.distplot(df3['FMR'], hist=False, color="g", kde_kws={"shade": True})
```

Pros

    Shows percentage of each value with respect to the whole dataset
    Provides clearer image of where msot values lie between
    Slightly better at showing minimum and maximum values than histogram
    
Cons

    Does not show the population size

# Looking at US Population Change over Time

```python
df = (pd.ExcelFile('US_population.xlsx')).parse('Sheet1')
x = df['Year']
y = df['Natural_change']/1000000
y2 = df['Live_births']
y3 = df['Deaths']
df.head()
```

```python
sns.set_color_codes("pastel")

sns.barplot(x=x,y=y2,data=df,color='b')

sns.set_color_codes("muted")

sns.barplot(x=x, y=y3, data=df, color = 'b')
```

# Natural Change

```python
b1 = sns.barplot(x, y, data=df, color = 'b')
b1.set_ylabels("Natural Change (in Millions)")
```

Pros

    Shows change over time for each year
Cons
    
    Cluttered, lots of wasted space between 0 and 1

```python
bx = sns.pointplot(x, y, data=df, size = 1)
```

Pros

    Shows trendlines clearly
Cons


```python
sns.regplot(x, y, data = df, fit_reg = False)
```

Pros

    Least cluttered representation of population change
Cons
