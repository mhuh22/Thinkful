```python
import numpy as np
import pandas as pd
import scipy
from datetime import datetime
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_ar = [15]
for i in range(500):
    y = 5 + .7*x_ar[i-1] 
    x_ar.append(float(y))
```

```python
plt.plot(x_ar)
plt.title('Autoregressive Example')
plt.show()
```

```python
x_ar = [15]
for i in range(500):
    y = 5 + x_ar[i-1] 
    x_ar.append(float(y))
    
plt.plot(x_ar)
plt.title('Autoregressive Example: B = 1')
plt.show()
```

```python
x_ar = [15]
for i in range(500):
    y = 5 + 2*x_ar[i-1] 
    x_ar.append(float(y))
    
plt.plot(x_ar)
plt.title('Autoregressive Example: B = 2')
plt.show()
```

# Moving Average

```python
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + .3*e_t + e
    x_ma.append(float(y))
    e_t=e
```

```python
plt.plot(x_ma)
plt.title('Moving Average Example')
plt.show()
```

```python
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + 1*e_t + e
    x_ma.append(float(y))
    e_t=e
    
plt.plot(x_ma)
plt.title('Moving Average Example')
plt.show()
```

# Finding the Order

```python
from statsmodels.tsa.stattools import acf, pacf
x_acf = pd.DataFrame(pacf(x_ar))
x_acf.plot(kind='bar')
plt.title("AR(1) PACF")
```

```python
x_acf = pd.DataFrame(pacf(x_ma))
x_acf.plot(kind='bar')
plt.title('MA(1) PACF')
```
