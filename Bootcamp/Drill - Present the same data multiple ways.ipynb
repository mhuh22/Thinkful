# Plotting market value

```python
df2 = pd.read_csv('thads2013n.txt')
df3 = df2[['BUILT',
                               'FMR',
                               'FMTBEDRMS',
                               'FMTOWNRENT',
                               'FMTREGION',
                               'PER',
                               'TOTSAL', 
                               'VALUE',
                               'ZINC2',
                               'ZSMHC']]
df3.head()
```

```python
sns.set_style("whitegrid")
ax = sns.boxplot(y=df3['FMR'], data=df3)
```

Pros

    Easy to find median, quartiles, and outliers
    Shows where most values lie between
    Can show distribution skew with the number of outliers
    
Cons

    Hard to visualize outliers if there are a lot of them
    Doesn't show population size

```python
sns.violinplot(df3['FMR'], vert=False)
```

Pros

    Shows where most common values lie
    Shows median and quartiles
Cons

    Doesn't include population size or distribution
    Can't see minimum and maximum values

```python
sns.distplot(df3['FMR'], kde=False)
```

Pros 

    Best visualization for showing distribution 
    Shows most common values, and outliers
    Shows quantity of units for each price
    
Cons

    Hard to visualize the minimum and maximum values

```python
sns.distplot(df3['FMR'], hist=False, color="g", kde_kws={"shade": True})
```

Pros

    Shows percentage of each value with respect to the whole dataset
    Provides clearer image of where msot values lie between
    Slightly better at showing minimum and maximum values than histogram
    
Cons

    Does not show the population size

# Flight Data (Continuous and categorical variable)

```python
flights = sns.load_dataset("flights")
print(flights.head(12))
```

```python
# Comparing months using boxplots
ax = sns.boxplot(x='month', y='passengers',data = flights)
```

Pros

    Presents min, quartiles, median, and max clearly
Cons

    Can't determine sample size

```python
# Comparing months using barplots
g = sns.factorplot(x='month', y='passengers', data = flights, kind='bar')
```

Pros

    Clearly shows averages for each month, and margin of error
Cons

    Large amount of wasted space between 0 and 200
    Margins of error are large due to large sample size

```python
plt.figure(figsize=(12,8))

# Comparing months using pointplots
g = sns.factorplot(x='month', y='passengers', data = flights, kind='point')
```

Pros

    Best at showing margins of error
    Shows clear difference between July/ August and the rest
Cons

    Can be misleading since it doesn't start at 0

```python
plt.figure(figsize=(12,8))

# Comparing months using violinplots
ax = sns.violinplot(x='month', y='passengers', data = flights, kind='point')
```

Pros

    Presents min, quartiles, median, and max clearly
    Shows which number of passengers is more common
Cons

    Harder to see median than others, and gauge sample size

```python
# Comparing months using a stripplot
sns.stripplot(x='month', y='passengers',data = flights)
```

Pros

    Min, max, range, and spread are clear
    Easy to identify sample size (at least for this dataset)
Cons

    Hard to visualize any overlap
    Difficult to identify median, mean, mode

```python
# Comparing months using a swarmplot
sns.swarmplot(x='month', y='passengers',data = flights)
```

Pros

    Min, max, range, and spread are clear
    Easy to visualize any overlap
    Easy to identify sample size (at least for this dataset)
Cons

    Difficult to identify median, mean
