```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Grab and process the raw data.
data_path = ("C:\\Users\\mhuh22\\Desktop\\Thinkful\\Unit_2\\2.2. Bayes Model\sentiment labelled sentences\\imdb_labelled.txt")
sms_raw = pd.read_csv(data_path, delimiter= '\t', header=None)
sms_raw.columns = ['message','rating']
```

```python
sms_raw.loc[sms_raw['rating']==0]
```

```python
keywords = ['horrendous','bad','not','predictable','awful','worst','hated','boring','disappointing','garbage']
keywords = ['']
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    sms_raw[str(key)] = sms_raw.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
```

```python
data = sms_raw[keywords]
target = sms_raw['rating']

# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```
