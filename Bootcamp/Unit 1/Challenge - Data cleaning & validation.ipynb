```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('C:\\Users\\mhuh22\\Desktop\\Thinkful\\Unit_1\\3. Intermediate Visualization\\')
journals= pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding='latin1')
```

```python
journals.head()
```

```python
# cleaning up the data

# drop rows with na values
journals = journals.dropna()

# capitaize journal names
journals['Journal title'] = journals['Journal title'].apply(lambda x: str(x).capitalize())

# cleaning up publisher section
journals['Publisher'] = journals['Publisher'].str.capitalize()
journals['Publisher'] = journals['Publisher'].str.strip()
journals['Publisher'] = journals['Publisher'].str.replace(r'/|\(|\)|','')

# abbreviate 'cost' column
journals = journals.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)': 'cost'})

# deletes currency symbols and converts the cost into a manipulable float
journals['cost'] = journals['cost'].str.replace(r'£|\$','',).astype(float)

# removes rows where the cost is 999999. which clearly is bad data
journals = journals[journals.cost != 999999.00]

# adds counter that will be used for counting journals
journals['count'] = 0
```

```python
# count appearances of each journal
journal_count = journals[['Journal title','count']].groupby(['Journal title']).count()

# sort and return top 5 most common journals
(journal_count.sort_values(['count'], ascending=False)).head()
```

```python
journals.drop(['count'],axis=1)

journals['Publisher'] = journals['Publisher'].str.title()
journals['Publisher'] = journals['Publisher'].replace('\t','')

journals['Publisher'] = journals['Publisher'].replace(['Acs','Acs Amercian Chemical Society',
                                                       'American Chemical Society Publications',
                                                       'Acs Amercian Chemical Society Publications',
                                                       'Acs Publications'],'American Chemical Society')
journals['Publisher'] = journals['Publisher'].replace(['Ambsb','American Society For Biochemistry And Molecular Biolgy',
                                                       'American Soc For Biochemistry And Molecular Biology',
                                                       'Asbmbcenveo Publisher Services',
                                                       'Asbmb Cadmus','Asbmbcadmus'
                                                       ],'American Society for Biochemistry and Molecular Biology')
journals['Publisher'] = journals['Publisher'].replace(['Asm','Asm American Society For Microbiology',
                                                       'American Society Of Microbiology	'
                                                       ],'American Society For Microbiology')
journals['Publisher'] = journals['Publisher'].replace(['American Society Of Hamatology'],'American Society of Hematology')
journals['Publisher'] = journals['Publisher'].replace('Benthan Science Publishers','Bentham Science Publishers')
journals['Publisher'] = journals['Publisher'].replace(['Biomed Central Ltd','Biomed Central'],'Biomed Central Limited')
journals['Publisher'] = journals['Publisher'].replace(['Bmj','Bmj Journals','Bmj Group','Bmj Publishing Group'],
                                                      'British Medical Journal')
journals['Publisher'] = journals['Publisher'].replace('Cadmus','Cadmus Journal Service')
journals['Publisher'] = journals['Publisher'].replace(['The Company Of Biolgists','The Company Of Biologists Ltd'],'The Company Of Biologists	')
journals['Publisher'] = journals['Publisher'].replace('The Endocrine Socety','The Endocrine Society')

journals['Publisher'] = journals['Publisher'].replace(['Wiley Subscription Serviices Inc',
                                                       'Wiley Subscription Services Inc.',
                                                      'Wiley Subscription Serviices Inc'],
                                                     'Wiley Subscription Services')
journals['Publisher'] = journals['Publisher'].replace(['Wliey-Blackwell','Wileyblackwell'],'Wliey-Blackwell')
journals.groupby(['Publisher']).mean()
```
